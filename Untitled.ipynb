# NLP - Natural Processing Language

```python
def remove_punctuation(txt):
    import string
    return txt.translate(str.maketrans("","",string.punctuation)).lower()
```

```python
data["Review_that_there_is_no_punctuation"]=data["Reviews"].apply(remove_punctuation)
```

# Removing The Stop Words

```python
import nltk.corpus
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
stop_words=stopwords.words("english")
print(sorted(stop_words))
```

```python
len(stop_words)
```

```python
data["Review_Clean"]=data["Review_that_there_is_no_punctuation"].apply(
    lambda x:" ".join([word for word in x.split() if word not in (stop_words)])
)
data
```

```python
not_stop_words=["not","no"]
stop_words_new=[x for x in stop_words if x not in not_stop_words]
print(sorted(stop_words_new))
```

```python
data["Review_Clean"]=data["Review_that_there_is_no_punctuation"].apply(
    lambda x:" ".join([word for word in x.split() if word not in (stop_words_new)])
)
data
```

```python
data_tfidf=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
data_tfidf["Sentiment"]=data["Sentiment"]
data_tfidf["Reviews"]=data["Reviews"]
data_tfidf["Review_Clean"]=data["Review_Clean"]
data_tfidf
```

# Sentiment Analysis
We will try to solve this problem by the help of these methods
1. Logistic Regression
2. Random Forest 
3. Neural Networks

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(data_tfidf,test_size=0.2,random_state=1)
```

```python
print(train.shape)
print(test.shape)
print(data_tfidf.shape)
```

# By the help of Logistic Regression

```python
#First we will be Training the Model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
features=vectorizer.get_feature_names_out()
target="Sentiment"
```

```python
model_Logistic=LogisticRegression(
    solver="newton-cg",
    max_iter=100,
    verbose=2,
    n_jobs=-1,
    random_state=1
).fit(train[features],train[target])
```

# Predictions

```python
test["Prediction_Logistic"]=model_Logistic.predict(test[features])
```

```python
test[["Sentiment","Prediction_Logistic","Reviews","Review_Clean"]].sample(20,random_state=2)
```

# Metrics

```python
from sklearn.metrics import accuracy_score , f1_score
```

```python
print("-"*75)
print("Logistic regression scores:")
print("Accuracy score:",accuracy_score(test[target],test["Prediction_Logistic"]))
print("F1 Score:",f1_score(test[target],test["Prediction_Logistic"]))
print("-"*75)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier(
n_estimators=500,
max_depth=25,
random_state=1).fit(train[features],train[target])
```

```python
test["Prediction_RF"]=model_rf.predict(test[features])
test[["Sentiment","Prediction_Logistic","Prediction_RF","Reviews","Review_Clean"]].sample(20,random_state=2)
```

```python
print("-"*75)
print("Random Forest Scores:")
print("Accuracy score:",accuracy_score(test[target],test["Prediction_RF"]))
print("F1 Score:",f1_score(test[target],test["Prediction_RF"]))
print("-"*75)
```

# Neural Networks

```python
from sklearn.neural_network import MLPClassifier
model_nn=MLPClassifier(
hidden_layer_sizes=(50,50,50),
solver="lbfgs",
max_iter=1000,
activation="relu",
random_state=1).fit(train[features],train[target])
test["Prediction_MLP"]=model_nn.predict(test[features])
test[["Sentiment","Prediction_Logistic","Prediction_RF","Prediction_MLP","Reviews","Review_Clean"]].sample(20,random_state=2)
index=874
print(test[["Reviews"]].loc[index].values)
print(test[["Review_Clean"]].loc[index].values)
print("-" * 75)
print("Logistic regression scores:")
print("Accuracy score:",accuracy_score(test[target],test["Prediction_Logistic"]))
print("F1 Score:",f1_score(test[target],test["Prediction_Logistic"]))
print("-" * 75)
print("-" * 75)
print("Logistic regression scores:")
print("Accuracy score:",accuracy_score(test[target],test["Prediction_RF"]))
print("F1 Score:",f1_score(test[target],test["Prediction_RF"]))
print("-" * 75)
print("-" * 75)
print("Neural Network Scores:")
print("Accuracy score:",accuracy_score(test[target],test["Prediction_MLP"]))
print("F1 Score:",f1_score(test[target],test["Prediction_MLP"]))
print("-" * 75)
```
